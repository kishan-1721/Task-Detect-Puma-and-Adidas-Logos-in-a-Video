```python
save_video_path = r"C:\Users\kisha\Projects\YOLO\Brand_Logo_Detection\Task_Mobile_Net_Project\Testing_Video.mp4"

# --------------------------------------------------------
# 2. LOAD TRAINED YOLOv11 MODEL
# --------------------------------------------------------
model = YOLO(r"C:\Users\kisha\Projects\YOLO\Brand_Logo_Detection\Task_Mobile_Net_Project\runs\detect\train\weights\best.pt")

# --------------------------------------------------------
# 3. PROCESS VIDEO FRAME BY FRAME
# --------------------------------------------------------
cap = cv2.VideoCapture(save_video_path)

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
output_path = r"C:\Users\kisha\Projects\YOLO\Brand_Logo_Detection\Task_Mobile_Net_Project\Output\output_labeled_video.mp4"
out = cv2.VideoWriter(output_path, fourcc, 30,
                      (int(cap.get(3)), int(cap.get(4))))

# --------------------------------------------------------
# 4. CSV FILE SETUP
# --------------------------------------------------------
csv_path = r"C:\Users\kisha\Projects\YOLO\Brand_Logo_Detection\Task_Mobile_Net_Project\Output\detections.csv"

csv_file = open(csv_path, mode='w', newline='')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["frame", "class", "confidence", "xmin", "ymin", "xmax", "ymax"])

# --------------------------------------------------------
# 5. RUN DETECTION LOOP
# --------------------------------------------------------
frame_no = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_no += 1

    results = model(frame, conf=0.45, stream=True)

    for r in results:
        for box in r.boxes:
            cls = int(box.cls[0])
            conf = float(box.conf[0])
            xmin, ymin, xmax, ymax = map(int, box.xyxy[0])

            # Get label name from model (ex: "Adidas", "Puma")
            class_name = model.names[cls]

            # Draw bounding box
            cv2.rectangle(frame, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2)
            cv2.putText(frame, f"{class_name} {conf:.2f}",
                        (xmin, ymin - 10),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.7, (0, 255, 0), 2)

            # WRITE TO CSV
            csv_writer.writerow([frame_no, class_name, conf, xmin, ymin, xmax, ymax])

    out.write(frame)

cap.release()
out.release()
csv_file.close()

print("Video saved at:", output_path)
print("CSV saved at:", csv_path)
```
